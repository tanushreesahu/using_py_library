## Seaborn Library

Seaborn line plot is python data visualization lirary based on matplotib .Its provides informative statistical graphics

Matplotlib is better for basic plots and seaborn is better for more advanced statistical plots

```python
import seaborn as sns
```

```python
import matplotlib.pyplot as plt
```

```python
# To view all available data sets in seaborn 
# get_dataset_names()

print(sns.get_dataset_names())
```

```python
# using fmri data set
# load the data set

fmri=sns.load_dataset("fmri")
```

```python
fmri.head()
```

```python
fmri.shape
```

```python
sns.lineplot(x="timepoint",y='signal',data=fmri)
plt.show()
```

```python
sns.lineplot(x="timepoint",y='signal',data=fmri,hue='event')
plt.show()
```

```python
# adding styles

sns.lineplot(x='timepoint',y='signal',data=fmri,hue='event',style='event')
plt.show()
```

```python
# adding markers

sns.lineplot(x="timepoint",y='signal',data=fmri,hue='event',style='event',markers=True)
plt.show()
```

# Jointplot

Two geometric combination scatter + histrogram

```python
import seaborn as sns
import matplotlib.pyplot as plt

iris=sns.load_dataset('iris')
```

```python
iris.head()
```

```python
sns.jointplot(x='sepal_length', y='petal_length',data=iris)
plt.show()
```

```python
# color change 

sns.jointplot(x='sepal_length',y='petal_length',data=iris, color='olive')
plt.show()
```

```python
# kind parameter= we can add regression line 
```

```python
sns.jointplot(x='sepal_length',y='petal_length',data=iris,color='pink',kind='reg')
plt.show()
```

# Boxplot

when multivariate used

Normally used where numerical and categories value represent.

categories variable represented on x axis

NUmerical variable represented on y axis

```python
import pandas as pd
```

```python
print(sns.get_dataset_names())
```

```python
churn=pd.read_csv("Customer-Churn.csv")
```

```python
churn.head()
```

```python
sns.boxplot(x='Churn', y='tenure',data=churn)
plt.show()
```

```python
sns.boxplot(x='InternetService', y='MonthlyCharges',data=churn)
plt.show()
```

```python
# palette /color
```

```python
sns.boxplot(x='Contract',y='tenure',data=churn,palette='deep')
plt.show()
```

```python
sns.boxplot(x='Contract',y='tenure',data=churn,linewidth=4)
plt.show()
```

```python
sns.boxplot(x='Contract',y='tenure',data=churn,order=['Two year','Month-to-month','One year'])
plt.show()
```

```python
sns.boxplot(x='Contract',y='tenure',data=churn,hue='PaymentMethod')
plt.show()
```
